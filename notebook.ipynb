```python
%matplotlib inline
import numpy as np
import pandas as pd
from pathlib import Path
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.manifold import TSNE

from matplotlib import pyplot as plt
from sklearn.datasets import make_regression
```

```python
# Import the data
file = Path('Resources/myopia.csv')
df = pd.read_csv(file)
df
```

```python
target = df['MYOPIC']
features = df.drop('MYOPIC', axis=1)
features.head
```

```python
scaler = StandardScaler()
df_scaled = scaler.fit_transform(features)
```

```python

# Initialize PCA model
pca = PCA(n_components=0.9)

# Get two principal components for the iris data.
myopia_pca = pca.fit_transform(df_scaled)
```

```python
myopia_pca.shape
```

```python
# Fetch the explained variance
pca.explained_variance_ratio_
```

```python
pca.explained_variance_ratio_.sum()
```

```python
tsne = TSNE(learning_rate="auto", init="pca", n_iter= 250)
tsne_features=tsne.fit_transform(myopia_pca)
```

```python
tsne_features.shape
```

```python
plt.scatter(tsne_features[:,0], tsne_features[:,1])
plt.show()
```

```python
plt.scatter(tsne_features[:,0], tsne_features[:,1], c=target)
plt.show()
```

```python
cluster_myopia_pca = pd.DataFrame(
    data=myopia_pca, columns=['m1', 'm2','m3','m4','m5','m6','m7','m8','m9','m10'])
cluster_myopia_pca.head()
```

```python
inertia = []
k = list(range(1, 11))

for i in k:
    km = KMeans(n_clusters=i, random_state=0)
    km.fit(cluster_myopia_pca)
    inertia.append(km.inertia_)

# Define a DataFrame to plot the Elbow Curve using hvPlot
elbow_data = {"k": k, "inertia": inertia}
df_elbow = pd.DataFrame(elbow_data)

plt.plot(df_elbow['k'], df_elbow['inertia'])
plt.xticks(range(1,11))
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
def get_clusters(k, data):
    # Initialize the K-Means model
    model = KMeans(n_clusters=k, random_state=0)

    # Train the model
    model.fit(data)

    # Predict clusters
    predictions = model.predict(data)

    # Create return DataFrame with predicted clusters
    data["class"] = model.labels_

    return data
```

```python
clusters = get_clusters(3, cluster_myopia_pca)
clusters
```

```python
def show_clusters(cluster_myopia_pca):
    plt.scatter(cluster_myopia_pca['m1'], cluster_myopia_pca['m2'], c=cluster_myopia_pca['class'])
    plt.title('Clusters')
    plt.xlabel('m1')
    plt.ylabel('m2')
    plt.show()
```

```python
show_clusters(clusters)
```
